```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from scipy import stats
```

```python
data=pd.read_csv('Students.csv')
print('DATA BEFORE')
data.head()
```

```python
math_mean =data['MathScore'].mean()
eng_mean = data['EnglishScore'].mean()

data['EnglishScore'].fillna(eng_mean)
data['MathScore'].fillna(math_mean)

print("DATA AFTER HANDLING MISSING COLUMNS")
data.head()
```

```python
le = LabelEncoder()

data['Gender'] = le.fit_transform(data['Gender'])
```

```python
data.head()
```

```python
data['TotalScore']=data['EnglishScore']+data['MathScore']
```

```python
data.head()
```

```python
data['EnglishScore_z'] = stats.zscore(data['EnglishScore'])
data['MathScore_z'] = stats.zscore(data['MathScore'])
data['TotalScore_z'] = stats.zscore(data['TotalScore'])
```

```python
data.to_csv('student_solution_31.csv',index=False)
```
