```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from scipy import stats
```

```python
data = pd.read_csv('Sales.csv')
data.head()
```

```python
data['Revenue'] = data['Price'] * data['QuantitySold']
data.head()
```

```python
z_scores = stats.zscore(data['Price'])
data =data[(z_scores > -3) & (z_scores < 3)]
data.head()
```

```python
scaler = MinMaxScaler()
data[['Price', 'Revenue']] = scaler.fit_transform(data[['Price', 'Revenue']])
data.head()
```

```python
data.to_csv('Sales__solution_31.csv',index=False)
```
