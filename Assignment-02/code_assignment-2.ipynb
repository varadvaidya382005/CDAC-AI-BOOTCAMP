```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Housing.csv')
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.describe()
```

```python
#Bar chart
plt.bar(df['bathrooms'],df['bedrooms'])
```

```python
#scatter graph
plt.scatter(df['bedrooms'],df['bathrooms'])
```

```python
#pie chart
plt.pie(df['parking'])
```

```python
#Histogram
plt.hist(df['price'])
```

```python
#count plot
plt.figure(figsize=(15,5))
sns.countplot(x='price',data=df)
```

```python
#Spliting according to furnishingstatus
plt.figure(figsize=(15,10))
sns.countplot(x='area',data=df,hue='furnishingstatus')
```

```python
#heatmap
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Compute the correlation matrix
corr_matrix = numeric_df.corr()

# Plot the heatmap
plt.figure(figsize=(10,5))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
#Box Plot
sns.boxplot(y='area',data=df)
```
